```python
## Linear Regression
#One-hot encoding used

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
#Bangalore
#one hot enncoding location
X = pd.get_dummies(bangalore[['area', 'location']], drop_first=True)
y = bangalore['price']

#train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Training the model
model = LinearRegression()
model.fit(X, y)

#Predict and evaluate
y_pred = model.predict(X_test)
print("Bangalore data Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Bangalore data R-squared Score:", r2_score(y_test, y_pred))
print("Bangalore data RMSE:", np.sqrt(mean_squared_error (y_test, y_pred)))
```

```python
#Bangalore
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Selecting features
clustering_data = bangalore[['price_per_sqft', 'area', 'price', 'no._of_bedrooms']].copy()

# Scaling features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# Applying KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
clustering_data['Cluster'] = kmeans.fit_predict(scaled_data)

# Visualize
plt.figure(figsize=(8, 5))
plt.scatter(clustering_data['area'], clustering_data['price_per_sqft'], c=clustering_data['Cluster'], cmap='viridis')
plt.xlabel("Area (sqft)")
plt.ylabel("Price per sqft")
plt.title("KMeans Clustering of Bangalore Properties")
plt.colorbar(label='Cluster ID')
plt.show()
```

```python
#Using elbow method to find best k value
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Standardize data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# Calculating best k value
sse = []
for k in range(1, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_data)
    sse.append(km.inertia_)

# Plot
plt.plot(range(1, 10), sse, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("SSE")
plt.show()

# hence k=4 is the best value
```
